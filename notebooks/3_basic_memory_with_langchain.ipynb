```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.2:3b",
    temperature=0.7
)
```

```python
prompts = [
    {"role":"user","content":"Hello, My name is Raghvendra Singh"},
    {"role":"ai","content":"Hello, Raghvendra Singh! How can I assist you today?"},
    {"role":"user","content":"What is my name?"}
]

res = llm.invoke(prompts)
```

```python
res.content
```

### Manage history

```python
history = []
while True:
    query = input("User: ")
    
    if query.lower() in ["exit","quit","bye"]:
        print("Good Bye ðŸ‘‹")
        break
    
    history.append({"role":"user","content":query})
    print("User: ",query)
    
    res = llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    
    print("AI: ",res.content, '\n')
```
